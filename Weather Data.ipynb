# Weather Data

In this notebook, we'll get some open weather data from http://climate.weather.gc.ca/, and use some basic functions of the python library called pandas, which contains tools for data analysis.

The data we'll use is daily weather data for September 2018 from a weather station in Vancouver, which can be found [here](http://climate.weather.gc.ca/climate_data/daily_data_e.html?hlyRange=1976-01-20%7C2018-09-25&dlyRange=1925-11-01%7C2018-09-24&mlyRange=1925-01-01%7C2007-02-01&StationID=888&Prov=BC&urlExtension=_e.html&searchType=stnProx&optLimit=yearRange&Month=9&Day=24&StartYear=1840&EndYear=2018&Year=2018&selRowPerPage=25&Line=2&txtRadius=25&optProxType=city&selCity=49%7C17%7C123%7C8%7CVancouver&selPark=&txtCentralLatDeg=&txtCentralLatMin=0&txtCentralLatSec=0&txtCentralLongDeg=&txtCentralLongMin=0&txtCentralLongSec=0&timeframe=2)
.

First, we'll use pandas to scan the page for the html table with the data. Running the cell below will display the raw data as a table. Since we'll be having the notebook get the data directly from the website, the data will be up to date for the time you run this notebook. As you will see, the raw data table is a bit messy; we'll need to choose what we want to display.

```python
#Setting up the tools we'll be using in this notebook:
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

#Scanning the web page for the html table:
tables = pd.read_html("http://climate.weather.gc.ca/climate_data/daily_data_e.html?hlyRange=1976-01-20%7C2018-09-25&dlyRange=1925-11-01%7C2018-09-24&mlyRange=1925-01-01%7C2007-02-01&StationID=888&Prov=BC&urlExtension=_e.html&searchType=stnProx&optLimit=yearRange&Month=9&Day=24&StartYear=1840&EndYear=2018&Year=2018&selRowPerPage=25&Line=2&txtRadius=25&optProxType=city&selCity=49%7C17%7C123%7C8%7CVancouver&selPark=&txtCentralLatDeg=&txtCentralLatMin=0&txtCentralLatSec=0&txtCentralLongDeg=&txtCentralLongMin=0&txtCentralLongSec=0&timeframe=2")

#Setting the table to a variable:
data = tables[0]

#Displaying the data table:
display(data)
```

### Graphing the data

If we want to graph the mean daily temperatures, we'll need to grab the appropriate columns from the dataframe above and use the matplotlib library to plot them:

```python
%matplotlib inline

#Converting the first and fourth columns to numeric data (missing data will be skipped and relabeled to "NaN"):
x = pd.to_numeric(data.iloc[:,0], errors='coerce')
y = pd.to_numeric(data.iloc[:,3], errors='coerce')

#Creating the plot:
fig, ax = plt.subplots(figsize=(20, 10))
plt.plot(x,y)

#Adding labels, tick marks, etc:
plt.title('Daily Mean Temperatures for September 2018', fontsize=18)
plt.xlabel('Day', fontsize=15)
plt.ylabel('Mean Temperature (°C)', fontsize=15)
#plt.ylim((0,25))
plt.xticks(np.arange(1, 31, step=1))

#Displaying the plot:
plt.show()
```
